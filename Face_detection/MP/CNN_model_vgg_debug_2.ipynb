```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import cv2
import os
import random
from PIL import Image
from skimage import io, filters
from scipy import ndimage
import tensorflow as tf
from keras.applications.vgg19 import VGG19, preprocess_input
from keras.models import Sequential, Model
from keras.layers import Dense, Dropout, Activation, Flatten, BatchNormalization, Conv2D, MaxPooling2D, GlobalAveragePooling2D
from keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from keras.preprocessing.image import ImageDataGenerator
```

```python
random.seed(seed)
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# Input shape required by VGG19
IMAGE_WIDTH=512
IMAGE_HEIGHT=512
IMAGE_CHANNELS=3
INP_SHAPE = (IMAGE_HEIGHT, IMAGE_WIDTH, IMAGE_CHANNELS)
IMAGE_SIZE=(IMAGE_HEIGHT, IMAGE_WIDTH)
BATCH_SIZE = 64
```

```python

# Load the VGG19 model without the top classification layers
vgg_model = VGG19(weights='imagenet', include_top=False, input_shape=INP_SHAPE)

# Freeze the layers of VGG19 model to not train them during the fine-tuning
for layer in vgg_model.layers:
    layer.trainable = False

# Create a new top model to put on top of the VGG19
x = vgg_model.output
x = Flatten()(x)  # Flatten the output of the conv layers

x = Dense(128)(x)
x = Activation('relu')(x)
x = Dropout(0.7)(x)

# Add a final output layer for binary classification with a sigmoid activation
predictions = Dense(1, activation='sigmoid')(x)

# This is the new model we will train
model = Model(inputs=vgg_model.input, outputs=predictions)
```

```python
model.summary()
```

```python
vggmodel_weights_filepath = 'vggmodel_best_debug_2.hdf5'
if os.path.exists(vggmodel_weights_filepath):
    model.load_weights(vggmodel_weights_filepath)
```

```python
from keras.optimizers import Adam
# Set a custom learning rate
learning_rate = 0.0001
```

```python
# Compile the model
model.compile(optimizer=Adam(learning_rate=learning_rate),
              loss="binary_crossentropy",
              metrics=[tf.keras.metrics.BinaryAccuracy(name='accuracy'),
                       tf.keras.metrics.Precision(name='precision'),
                       tf.keras.metrics.Recall(name='recall'),
                       ])
```

```python
callbacks = []

# Saving checkpoints
checkpoint = ModelCheckpoint(vggmodel_weights_filepath,
                             monitor='val_loss',
                             verbose=1,
                             save_best_only=True,
                             mode='min')
callbacks.append(checkpoint)

# Define early_stopping.
early_stopping = EarlyStopping(monitor='val_loss',
                               patience=8,
                               verbose=1,
                               restore_best_weights=True)
callbacks.append(early_stopping)

# Define the ReduceLROnPlateau callback
reduce_lr = ReduceLROnPlateau(monitor='val_loss',
                              factor=0.1,
                              patience=5)
callbacks.append(reduce_lr)
```

```python
#set a seed
seed = 72
np.random.seed(seed)
```

```python
# Concatenate X_train and y_train
train_df = pd.concat([X_train, y_train], axis=1)
# Concatenate X_val and y_val
val_df = pd.concat([X_val, y_val], axis=1)
# Concatenate X_test and y_test
test_df = pd.concat([X_test, y_test], axis=1)
```

```python
# Data augmentation and preprocessing
def apply_data_augmentation(df, batch_size=32):

    datagen = ImageDataGenerator(preprocessing_function=preprocess_input,
                                 rotation_range=15,
                                 brightness_range=[0.6, 1.0],
                                 horizontal_flip=True)

    image_generator = datagen.flow_from_dataframe(dataframe=df,
                                                  x_col='img_path',
                                                  y_col='label',
                                                  seed=seed,
                                                  # all images will be resized to configured target size,
                                                  target_size=IMAGE_SIZE,
                                                  batch_size=batch_size, # Default 32 images from this label
                                                  #interpolation='bicubic',
                                                  class_mode=None, # Setting this to None as we are doing class-wise augmentation.
                                                  )

    return image_generator
```

```python
# Filter autistic class images.
autistic_df = train_df[train_df['label'] == 'Autistic']
non_autistic_df = train_df[train_df['label'] == 'Non_Autistic']
```

```python
autistic_train_generator = apply_data_augmentation(autistic_df, batch_size=BATCH_SIZE//2)
non_autistic_train_generator = apply_data_augmentation(non_autistic_df, batch_size=BATCH_SIZE//2)
```

```python
def visualize_batch(image_generator, labels_included=False,
                    num_rows=4, num_cols=8,
                    bgr_to_rgb=False, reverse_preprocess=False):

    if labels_included:
        images, labels = next(image_generator)
    else:
        images = next(image_generator)

    if reverse_preprocess:
        # The mean values for VGG16 trained on ImageNet data
        mean = np.array([123.68, 116.779, 103.939], dtype=np.float32)

        # The preprocess_input for VGG16 subtracts the mean of each channel.
        # Add the mean to their corresponding channels
        images += mean.reshape((1, 1, 1, 3))


    if bgr_to_rgb: # VGG16 preprocess_input changes RGB images to BGR images.
        images = images[..., ::-1]  # Convert BGR to RGB

    # Initialize a grid for the images
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 20))
    axes = axes.flatten()

    for i in range(num_rows * num_cols):
        # Check if the index is within the batch size
        if i < images.shape[0]:
            img = images[i]
            # Clip and convert to uint8
            img = np.clip(img, 0, 255).astype('uint8')

            if labels_included:
                label = "Autistic" if labels[i] == 1 else "Non_Autistic"
                axes[i].set_title(f'Label: {label}')

            axes[i].imshow(img)
            axes[i].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
# Visualizing the autistic images in BGR format
visualize_batch(autistic_train_generator)
```

```python
visualize_batch(non_autistic_train_generator)
```

```python
def combined_generator(autistic_generator, non_autistic_generator):
    while True:
        # Get a batch of images from the autistic_generator
        autistic_batch_x = next(autistic_generator)
        autistic_generator_filenames = autistic_generator.filenames

        # Get the batch size
        autistic_batch_size = autistic_batch_x.shape[0]
        # Autistic image labels are 1
        autistic_batch_y = np.ones((autistic_batch_size, 1))


        # Get a batch of images from the non-autistic generator
        non_autistic_batch_x = next(non_autistic_generator)
        non_autistic_generator_filenames = non_autistic_generator.filenames

        # Get the batch size
        non_autistic_batch_size = non_autistic_batch_x.shape[0]
        # Non-autistic image labels are 0
        non_autistic_batch_y = np.zeros((non_autistic_batch_size, 1))

        # Combine the batches
        batch_x = np.vstack((autistic_batch_x, non_autistic_batch_x))
        batch_y = np.vstack((autistic_batch_y, non_autistic_batch_y))
        batch_filenames = autistic_generator_filenames + non_autistic_generator_filenames

        # Shuffle the combined batch to avoid sequence of autistic and non-autistic images.
        # Total batch size should be 64
        indices = np.random.permutation(batch_x.shape[0])
        batch_x = batch_x[indices]
        batch_y = batch_y[indices]
        batch_filenames = [batch_filenames[i] for i in indices]

        yield batch_x, batch_y

# Combine the two generators
train_generator = combined_generator(autistic_train_generator,
                                     non_autistic_train_generator)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Prediction on test set
predictions = model.predict(test_generator)
predictions = predictions.flatten()  # Convert predictions to 1D array if necessary

# Apply 0.5 threshold
predictions_binary = np.where(predictions > 0.68, 1, 0)

# True Labels
# Ensure test_generator labels are in the same order as predictions
true_labels = test_generator.classes

# Compute Confusion Matrix
conf_matrix = confusion_matrix(true_labels, predictions_binary)

# Extract individual elements
TN, FP, FN, TP = conf_matrix.ravel()

# Create a new confusion matrix with labels
conf_matrix_labelled = np.array([[f"TN\n{TN}", f"FP\n{FP}"], [f"FN\n{FN}", f"TP\n{TP}"]])

# Plot Confusion Matrix
plt.figure(figsize=(6, 5))
ax = sns.heatmap(conf_matrix, annot=conf_matrix_labelled, fmt='', cmap='Blues', cbar=False)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
ax.set_xticklabels(['Negative', 'Positive'])
ax.set_yticklabels(['Negative', 'Positive'])
plt.show()
```

```python
from sklearn.metrics import classification_report

# Class names
class_names = ['Non_Autistic', 'Autistic']

# Generate classification report
report = classification_report(true_labels, predictions_binary, target_names=class_names)
print(report)
```
